feature extraction

```python
import mne

# feature_extraction.py에 작성된 함수들
def load_eeg_data(file_path):
    data = pd.read_csv(file_path)
    return data

def compute_band_power(raw, band):
    fmin, fmax = band  # 주파수 대역 설정
    data = raw.get_data()
    sfreq = raw.info['sfreq']
    data = data[:-1, :]  # 마지막 열은 라벨이므로 제외
    psds, freqs = mne.time_frequency.psd_array_welch(data, sfreq=sfreq, fmin=fmin, fmax=fmax, n_fft=256)  # 전력 스펙트럼 밀도 계산
    # 주파수 대역의 전력 계산
    band_power = np.sum(psds, axis=-1)
    return band_power

def extract_features(data, sfreq=250):
    eeg_data = data.values  # 모든 열을 가져옴
    ch_names = data.columns.tolist()  # 열 이름을 채널 이름으로 사용
    info = mne.create_info(ch_names=ch_names, sfreq=sfreq, ch_types='eeg')  # MNE의 Info 객체 생성
    raw = mne.io.RawArray(eeg_data.T, info)  # RawArray 객체 생성

    # 주파수 대역 정의
    delta_band = (0.5, 4)
    theta_band = (4, 8)
    alpha_band = (8, 12)

    # 주파수 대역별 전력 계산
    delta_power = compute_band_power(raw, delta_band)
    theta_power = compute_band_power(raw, theta_band)
    alpha_power = compute_band_power(raw, alpha_band)

    # 전력 비율 계산
    alpha_delta_ratio = np.mean(alpha_power / delta_power)
    theta_alpha_ratio = np.mean(theta_power / alpha_power)
    delta_theta_ratio = np.mean(delta_power / theta_power)

    # 피처 딕셔너리 생성
    features = {
        'Alpha:Delta Power Ratio': alpha_delta_ratio,
        'Theta:Alpha Power Ratio': theta_alpha_ratio,
        'Delta:Theta Power Ratio': delta_theta_ratio
    }

    return features


# Epoch 크기
epoch_size = 1000

# DataFrame을 epoch로 나누는 함수
def split_dataframe_to_epochs(eeg_d, epoch_size):
    epochs = []
    for start in range(0, len(eeg_d), epoch_size):
        epoch = eeg_d.iloc[start:start + epoch_size]
        if len(epoch) == epoch_size:  # 마지막 epoch가 epoch_size보다 작을 경우 제외
            epochs.append(epoch)
    return epochs

# DataFrame을 epoch로 나누기
epochs = split_dataframe_to_epochs(eeg_d, epoch_size)

# 각 epoch에서 61번 column을 제외한 DataFrame을 새로운 리스트에 추가
new_epochs = [epoch.drop(epoch.columns[60], axis=1) for epoch in epochs]

# 각 epoch마다 피처를 추출하여 리스트에 저장
all_features = []

for idx, epoch in enumerate(new_epochs):
    features = extract_features(epoch)
    all_features.append(features)
    print(f"Epoch {idx+1}/{len(new_epochs)} processed.")

# 모든 피처를 하나의 DataFrame으로 결합
all_features_df = pd.DataFrame(all_features)
```

```python
# 결과 확인
print(f"Total number of epochs: {len(new_epochs)}")
print(f"Shape of features DataFrame: {all_features_df.shape}")
print(all_features_df)
```

```python
all_features_df
```

```python
# 각 에포크의 길이 (rows)
epoch_size = 1000

# 에포크 수
num_epochs = eeg_d.shape[0] // epoch_size

# 각 에포크에서 라벨 값 추출
labels = []
for i in range(num_epochs):
    # 에포크 데이터 추출
    epoch_data = eeg_d.iloc[i * epoch_size : (i + 1) * epoch_size]
    # 라벨 값 추출 (라벨 열의 첫 번째 값 사용)
    label = epoch_data['label'].iloc[0]
    labels.append(label)

# 라벨 값으로 새로운 DataFrame 생성
labels_df = pd.DataFrame(labels, columns=['label'])

# 결과 출력
labels_df
```

```python
# 학습 세트와 테스트 세트로 나누기 (80% 학습, 20% 테스트)
# X_train, X_test, y_train, y_test = train_test_split(feature_df, labels_df, test_size=0.2, random_state=42)
X_train, X_test, y_train, y_test = train_test_split(all_features_df, labels_df, test_size=0.2, random_state=42)
```

ML Model - 1. RandomForest

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import f1_score, cohen_kappa_score

model = RandomForestClassifier(n_estimators=100, random_state=42)
```

```python
# 모델 학습
model.fit(X_train, y_train)
```

```python
# 예측
y_pred = model.predict(X_test)
```

```python
# 성능 평가
accuracy = accuracy_score(y_test, y_pred)
accuracy
```

```python
# F1 스코어 계산
f1 = f1_score(y_test, y_pred, average='weighted')
print(f"F1 Score: {f1}")

# Kappa 계수 계산
kappa = cohen_kappa_score(y_test, y_pred)
print(f"Cohen's Kappa Score: {kappa}")
```
