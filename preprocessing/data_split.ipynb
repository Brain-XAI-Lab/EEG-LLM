```python
import pandas as pd
```

```python
# Load the data
data = pd.read_csv('/Users/imdohyeon/Documents/PythonWorkspace/EEG-GPT/Dataset/eeg_data_180000.csv')
data
```

```python
"""
파이프라인

1. 데이터를 1000개 단위로 묶어 배열에 저장
2. 이 배열의 인덱스를 무작위로 섞음
3. 섞인 인덱스의 80%를 훈련 데이터로, 20%를 테스트 데이터로 사용
4. 묶어 둔 인덱스를 다시 데이터로 원상복구
"""

# 데이터를 1000개 단위로 묶어 배열에 저장
data_list = []
for i in range(0, 180000, 1000):
    data_list.append(data.iloc[i:i+1000])
data_list[151]
```

```python
# 이 배열의 인덱스를 무작위로 섞음
import random
random.shuffle(data_list)
```

```python
data_list[0]
```

```python
# 섞인 총 180개 인덱스의 80%를 훈련 데이터로, 20%를 테스트 데이터로 사용
train_data = pd.DataFrame()
test_data = pd.DataFrame()
for i in range(0, 144):
    train_data = pd.concat([train_data, data_list[i]])
for i in range(144, 180):
    test_data = pd.concat([test_data, data_list[i]])
```
