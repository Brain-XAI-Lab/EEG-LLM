```python
import csv_to_json as ctoj
import pandas as pd
import numpy as np
import json
```

```python
# Simulated example data
data = {
    'E1': np.random.randn(2500),
    'E9': np.random.randn(2500),
    'E13': np.random.randn(2500)
}
df = pd.DataFrame(data)
window_size = 500
selected_columns = ['E1', 'E9', 'E13']
labels = ['left', 'right', 'top', 'bottom', 'left'] * (len(df) // window_size)

json_result = ctoj.csv_to_json(df, window_size, selected_columns, labels)

# Displaying a sample output with 5 prompts
sample_output = json.dumps(json_result[:5], indent=4)
print(sample_output)
```
